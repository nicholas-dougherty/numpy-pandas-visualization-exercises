## 1.) Use matplotlib to plot the following equation:
        y = x**2 - x + 2
You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
print(plt)
# everything worked 
```

```python
a=[]
b=[]

for x in range(-50,50):
    y=x**2-x+2
    a.append(x)
    b.append(y)

fig= plt.figure()
axes=fig.add_subplot(111)
axes.plot(a,b)
plt.show()
```

```python
x = list(range(-50, 50))
y = [n**2-n+2 for n in x]

fig, ax = plt.subplots()

plt.plot(x, y)
plt.title('The Beginning of the End')
plt.xlabel('$x$')
plt.ylabel('$x^2-x+2$')
plt.grid()

# the following lines set the origin
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.annotate('origin', xy = (0,0))

plt.show()
```

```python
import math

numbers = range(1, 20)
def f(x1):
    return math.sqrt(x1)

x1 = list(range(1, 20))
y1 = [f(x1) for x1 in numbers]

fig, ax = plt.subplots()

plt.plot(x1, y1)
plt.title('Square Root of X')
plt.xlabel('$x$')
plt.ylabel('√x')
plt.grid()

# the following lines set the origin
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(numbers, y1)
plt.show()

##########################################################
# y = x^3 

import math

numbers = range(-3, 4)
def f(x2):
    return x2**3

x2 = list(range(-3, 4))
y2 = [f(x2) for x2 in numbers]

fig, ax = plt.subplots()

plt.plot(x2, y2)
plt.title('X Cubed')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.grid()

# the following lines set the origin
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(numbers, y2)
plt.show()

##########################################################
# y = 2x

import math

numbers = range(-20, 20)
def f(x3):
    return x3*2

x3 = list(range(-20, 20))
y3 = [f(x3) for x3 in numbers]

fig, ax = plt.subplots()

plt.plot(x3, y3)
plt.title('Double Vision')
plt.xlabel('$x$')
plt.ylabel('$2x$')
plt.grid()

# the following lines set the origin
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(numbers, y3)
plt.show()

##########################################################
# y = 1/(x + 1)

import math

numbers = range(1, 20)
def f(x4):
    return 1/(x4+1)

x4 = list(range(1, 20))
y4 = [f(x4) for x4 in numbers]

fig, ax = plt.subplots()

plt.plot(x4, y4)
plt.title('Slippery Slope')
plt.xlabel('$x$')
plt.ylabel('$1/(x+1$')
plt.grid()

# the following lines set the origin
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.plot(numbers, y4)
plt.show()
```
